```python
import matplotlib.pyplot as plt

x1 = [0, 125]
y1 = [125, 125]

x2 = [125, 250]
y2 = [125, 0]

x3 = [125, 125]
y3 = [0, 125]

plt.figure(figsize=(10, 4))

plt.plot(x3, y3, label='Cutoff value', color='#FF6400')
plt.plot(x1, y1, label='RUL', color='#1f77b4')
plt.plot(x2, y2, color='#1f77b4')

plt.xlim(0, 300)
plt.ylim(0, 140)

plt.legend()
plt.xlabel('Cycles')
plt.ylabel('RUL (cycles)')

plt.show()
```
